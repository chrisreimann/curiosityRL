```python
import gymnasium as gym
import minigrid
import numpy as np
from EpsilonGreedy import EpsilonGreedy
from simpleTabEnv import GridWorldEnv
from gymnasium.envs.registration import register
```

```python
model = EpsilonGreedy("GridWorld-v0", 50)
model.learn(50000, batchsize=100)
```

```python
modelIM = EpsilonGreedy("GridWorld-v0", 50, IM=True)
modelIM.learn(50000, batchsize=100)
```

```python
model.test(1000)
model.learningGraph()
```

```python
modelIM.test(1000)
modelIM.learningGraph()
```

```python
modelT = modelIM
env = gym.make("GridWorld-v0", render_mode="human", size=model.env.size)
observation, info = env.reset()

for _ in range(500):
    action = modelT.predict(observation)
    observation, reward, terminated, truncated, info = env.step(action)

    if terminated or truncated:
        observation, info = env.reset()

env.close()
```
